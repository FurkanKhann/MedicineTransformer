```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
```

Path to where model and tokenizer is save

```python
drive_model_path = "/content/drive/MyDrive/t5_medicine_model"
```

```python
model = T5ForConditionalGeneration.from_pretrained(drive_model_path)
tokenizer = T5Tokenizer.from_pretrained(drive_model_path)
```

Function to preprocess

```python
import torch

def predict(text):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Tokenize and move input to same device
    input_ids = tokenizer(text, return_tensors="pt").input_ids.to(device)

    # Generate output
    output_ids = model.generate(input_ids, max_length=128, decoder_start_token_id=tokenizer.pad_token_id)

    # Decode
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)
```

```python
ans=predict("Nitrogen")
```
